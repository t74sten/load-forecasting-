```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('dataset2_train.xlsx')
df.fillna(0, inplace=True)
df = df.round(decimals=2)
df = df.drop('date_time', 1);
```

```python
df1 = pd.read_excel('dataset2_test.xlsx')
df1 = df1.drop('date_time', 1);
df1 = df1.round(decimals=2)
```

```python
y = df.S
```

```python
X = df.drop(['S'], axis=1)
```

```python
y_test = df1.S
```

```python
X_test = df1.drop(['S'], axis=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN = KNeighborsRegressor(n_neighbors = 3, algorithm='auto', leaf_size=10, p=1)
```

```python
KNN.fit(X,y)
```

```python
predictionKNN = KNN.predict(X_test)
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2,2,1)
plt.plot(predictionKNN, label="PredictionKNN")
plt.plot(y_test, label="RealData")
plt.grid(True)
plt.legend()
plt.subplot(2,2,2)
plt.plot(predictionKNN, label="PredictionKNN")
plt.plot(y_test, label="RealData")
plt.xlim(0,50)
plt.grid(True)
plt.legend()
plt.subplot(2,2,3)
plt.plot(predictionKNN, label="PredictionKNN")
plt.plot(y_test, label="RealData")
plt.xlim(50,100)
plt.grid(True)
plt.legend()
plt.subplot(2,2,4)
plt.plot(predictionKNN, label="PredictionKNN")
plt.plot(y_test, label="RealData")
plt.xlim(100,150)
plt.grid(True)
plt.legend()
```
